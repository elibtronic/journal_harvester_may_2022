```python
from pprint import pprint
from habanero import Crossref
import random
import pandas as pd

cr = Crossref()
```

```python
#Build 'cursor of results'

res = cr.journals(ids = "0888-4773", works = True, cursor = "*", cursor_max = 1300, progress_bar = True)
sum([ len(z['message']['items']) for z in res ])
items = [ z['message']['items'] for z in res ]
items = [ item for sublist in items for item in sublist ]
```

```python
len(items)
```

```python
for a in items:
    print(a['DOI'])
```

```python
#One Random Entry to show fields in record

rando_art = random.choice(items)
pprint(rando_art)
```
